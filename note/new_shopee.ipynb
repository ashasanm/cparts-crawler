```python
import json
import csv
import time
from pprint import pprint

from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, InvalidArgumentException
```

```python
options = Options()
options.add_argument("--start-maximized")
options.add_argument('--user-agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36"')
options.add_argument('disable-infobars')
options.add_argument('--disable-extensions')
options.add_argument('--disable-blink-features=AutomationControlled')
driver = Chrome(executable_path="..\chromedriver\chromedriver.exe", options=options)
```

```python
driver.get("https://shopee.co.id/api/v4/search/search_items?by=relevancy&keyword=rx%206600&limit=60&newest=0&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2")
```

```python
shopee_json = driver.find_element("xpath", "/html/body/pre").text
```

```python
shopee_json = json.loads(shopee_json)
```

```python
def extract_data_from_json(json_item):
    item = json_item['item_basic']
    name = item['name']
    sold = item['sold']
    price = item['price']
    return {
        "title": name,
        "price": price,
        "sold": sold
    }

extracted_data = [extract_data_from_json(json_item) for json_item in shopee_json['items']]
extracted_data
```
